```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
countries = pd.read_csv("gapminder.csv")
countries.head(10)
```

### 0. Spočítej, kolik je v každém sloupci chybějících hodnot (countries.isnull().sum(axis = 0)). Na základě toho (a případně dalších úvah), vyber sloupce, který budeš pro PCA analýzu používat.

```python
countries.isnull().sum(axis = 0)
```

```python
# remove useless columns
countries_tidy = countries.drop(["iso", "year"], axis=1)
countries_tidy.sample(10)
```

```python
# find and hand replace missing value
countries_tidy[countries["income_groups"].isnull()]
```

```python
countries_tidy['income_groups'].fillna('high_income', inplace=True)
```

```python
# pick year only and replace missing values with 0
countries_tidy["eu_accession"] = pd.to_datetime(countries_tidy["eu_accession"])
countries_tidy['eu_accession'] = pd.DatetimeIndex(countries_tidy['eu_accession']).year
countries_tidy["eu_accession"].fillna(0, inplace=True)
countries_tidy.head()
```

```python
# pick year only and replace missing values with 0
countries_tidy["un_accession"] = pd.to_datetime(countries_tidy["un_accession"])
countries_tidy['un_accession'] = pd.DatetimeIndex(countries_tidy['un_accession']).year
countries_tidy["un_accession"].fillna(0, inplace=True)
countries_tidy.head()
```

```python
# NaN are replaced with related values
countries_tidy["life_expectancy_female"][3] = countries_tidy["life_expectancy"][3]
countries_tidy["life_expectancy_female"][48] = countries_tidy["life_expectancy"][48]
countries_tidy["life_expectancy_female"][106] = countries_tidy["life_expectancy"][106]
countries_tidy["life_expectancy_male"][3] = countries_tidy["life_expectancy"][3]
countries_tidy["life_expectancy_male"][48] = countries_tidy["life_expectancy"][48]
countries_tidy["life_expectancy_male"][106] = countries_tidy["life_expectancy"][106]
```

```python
countries_tidy.isnull().sum(axis = 0)
```

### 1. Do tabulky s vybranými sloupci doplň chybějící hodnoty pomocí SimpleImputer z balíčku sklearn.

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

```

```python
countries_tidy_imp = imp.fit_transform(countries_tidy[[
    "eu_accession",
    "area",
    "population",
    "alcohol_adults", 
    "bmi_men", 
    "bmi_women",
    "car_deaths_per_100000_people",
    "calories_per_day",
    "life_expectancy",
    "life_expectancy_female",
    "life_expectancy_male",
    "infant_mortality",
    "un_accession"
]])
```

```python
countries_tidy_imp = pd.DataFrame(countries_tidy_imp, columns=[
    "eu_accession",
    "area",
    "population",
    "alcohol_adults", 
    "bmi_men", 
    "bmi_women",
    "car_deaths_per_100000_people",
    "calories_per_day",
    "life_expectancy",
    "life_expectancy_female",
    "life_expectancy_male",
    "infant_mortality",
    "un_accession"], index=countries_tidy.name)
countries_tidy_imp.head()
```

### 2. Normalizuj data pomocí StandardScaler a spočítej první dvě komponenty PCA. Vykresli data do souřadnicového grafu (scatter plot). Jsou země EU blízko sebe? Tvoří shluk?

```python
# scale numerical data
scaler = StandardScaler()
countries_num_scaled = scaler.fit_transform(countries_tidy_imp)
countries_df_num_scaled = pd.DataFrame(countries_num_scaled,columns=countries_tidy_imp.columns)
countries_df_num_scaled
```

```python
# create PCA
pca = PCA(n_components=2)
countries_numeric_pca = pd.DataFrame(pca.fit_transform(countries_df_num_scaled), columns=['PCA1', 'PCA2'])
countries_numeric_pca
```

```python
# create a scatterplot
sns.scatterplot(countries_numeric_pca.PCA1, y=countries_numeric_pca.PCA2);
```

### 3. Dokážeš najít interpretaci pro PCA1 a PCA2 (první dvě komponenty PCA)? Pomoci může, když si nakreslíš interaktivní graf zobrazující jméno země (balíček plotly). Zkus si též spočítat, nakolik jednotlivé sloupečky z původní tabulky přispívají ke koeficientům (loadings) PCA1 a PCA2.

```python
import plotly.express as px
```

```python
fig = px.scatter(x=countries_numeric_pca.PCA1, y=countries_numeric_pca.PCA2,
                 color=countries_tidy_imp.life_expectancy, # PC1 explained
                 #color=countries_tidy_imp.bmi_men, # PC2 explained
                 hover_name = countries_tidy_imp.index)
fig.show()
```

### 4. Pokus se navrhnout, jak by šly do PCA zahrnout kategoriální veličiny, za předpokladu, že počet jejich kategorií není příliš velký (např. indikátor is_eu, zda je země v EU, nebo kontinent world_6region).

```python
# create dummy variables
countries_categorical_dv = pd.get_dummies(countries_tidy[["world_6region", "world_4region", "is_eu", "is_oecd"]])
countries_categorical_dv['name'] = countries_tidy.name
countries_categorical_dv.set_index("name", inplace=True)
countries_categorical_dv
```

```python
# join dataframes with categorical and numerical data
countries_catnum = countries_categorical_dv.join(countries_tidy_imp, how="outer")
countries_catnum
```

```python
# scale categorical and numerical data together
scaler = StandardScaler()
countries_catnum_scaled = scaler.fit_transform(countries_catnum)
countries_df_num_scaled = pd.DataFrame(countries_catnum_scaled,columns=countries_catnum.columns)
countries_df_num_scaled
```

```python
# create PCA
pca = PCA(n_components=2)
countries_catnum_pca = pd.DataFrame(pca.fit_transform(countries_catnum_scaled), columns=['PCA1', 'PCA2'])
countries_catnum_pca
```

```python
pca.explained_variance_ratio_
```

```python
fig2 = px.scatter(x=countries_catnum_pca.PCA1, y=countries_catnum_pca.PCA2, 
                 color=countries_tidy_imp.bmi_men,
                 #size=countries_tidy_imp.area,
                 #symbol=countries_tidy.world_4region,
                 hover_name = countries_catnum.index,
                 labels={
                     "x": "PC1 33%",
                     "y": "PC2 15 %",
                 })
fig2.show()

# clusters are formed based on world_4region category
# PC1 is explained by life_expectancy
# PC2 (unlike PC2 from numerical data only) is not explained by bmi_men, there are more factors contributing to PC2, not a clear distinction
```
