```python
def get_redundant_pairs(countires_corr):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = countires_corr.columns
    for sloupec in range(0, countires_corr.shape[1]):
        xfor j in range(0, sloupec+1):
            pairs_to_drop.add((cols[sloupec], cols[j]))
    return pairs_to_drop

def get_top_correlations(countires_corr, n=5):
    au_corr = countires_corr.unstack()
    labels_to_drop = get_redundant_pairs(countires_corr)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top Correlations")
print(get_top_correlations(countires_corr, 10))
```
