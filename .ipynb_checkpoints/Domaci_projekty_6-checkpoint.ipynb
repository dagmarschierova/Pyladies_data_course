## 0. Ve kterém roce/měsíci bylo nejvíc zrušených vlaků?

```python
import pandas as pd
```

```python
trains = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-02-26/full_trains.csv")
trains
```

```python
trains.dtypes
```

```python
trains["num_of_canceled_trains"].max()
```

```python
trains.loc[trains['num_of_canceled_trains'].idxmax()][["year", "month", "num_of_canceled_trains"]]
```

## 1. Která stanice je „nejhorší“? Např. dle počtu zrušených vlaků, největšího zpoždění na odjezdu apod.? Pozor, pro některé stanice jsou data jen z roku 2018.

```python
trains.loc[trains['num_of_canceled_trains'].idxmax()][["departure_station", "arrival_station", "num_of_canceled_trains"]]
```

```python
trains[trains.avg_delay_late_at_departure == trains.avg_delay_late_at_departure.max()][["departure_station", "arrival_station", "avg_delay_late_at_departure"]]
```

```python
trains[trains.avg_delay_all_arriving == trains.avg_delay_all_arriving.max()][["departure_station", "arrival_station", "avg_delay_all_arriving"]]
```

```python
trains.isna().sum()
```

## 2. Na které trati je průměrně největší zpoždění?

```python
trains[trains.avg_delay_late_at_departure == trains.avg_delay_late_at_departure.max()][["departure_station", "arrival_station", "avg_delay_late_at_departure"]]
```
